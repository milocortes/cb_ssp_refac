```python
# List of suffixes to remove

df = pd.read_csv('data/ssp_model_results/iran/ATTRIBUTE_STRATEGY_new.csv')

suffixes_to_remove = [
    "_LOW", "_LOWEST", "_HIGHEST", "_HIGHER", "_HIGH", "_LOWER",
    "_FROMTECH", "_URBPLAN"
]

# Function to remove suffixes from a single transformation specification string
def clean_transformation_specification(specification):
    transformations = specification.split('|')  # Split by |
    cleaned_transformations = [
        t.rstrip(''.join(suffix for suffix in suffixes_to_remove if t.endswith(suffix)))
        for t in transformations
    ]
    return '|'.join(cleaned_transformations)

# Apply the function to the `transformation_specification` column
df['transformation_specification'] = df['transformation_specification'].apply(clean_transformation_specification)

# Save the updated DataFrame to a new CSV file
df.to_csv('data/ssp_model_results/iran/ATTRIBUTE_STRATEGY_new.csv', index=False)

print("The file has been updated and saved as 'updated_file.csv'.")
```
