```python
df_out.info()
```

```python
strategy_codes = list(df_out.strategy_code.unique())
strategy_codes.sort()
strategy_codes
```

```python
# Iterate through unique strategy_code values
for strategy in strategy_codes:
    # Filter the DataFrame for the current strategy_code
    filtered_df = df_out[df_out["strategy_code"] == strategy]
    
    # Define the file name
    file_name = f"cost_benefit_results_{strategy}.csv"
    
    # Save to CSV
    filtered_df.to_csv(f'output/{file_name}', index=False)
    print(f"Saved {file_name}")
```
